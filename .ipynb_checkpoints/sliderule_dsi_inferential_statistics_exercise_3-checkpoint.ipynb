# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

# B

Using the author parameters I'm going to assume that small hospital has less than 1000 discharges, and the bigger ones, 1000 or more. 

# 1


Considering mean_small the mean excess ratio for the small hospital, and mean_large, the mean for the largers hospital excess ratio: <br><br>
<b>Null Hypothesis:</b> mean_small - mean_large = 0<br>
<b>Alternate Hypothesis:</b> mean_small - mean_large <> 0<br>


# 2


```python
data = clean_hospital_read_df

large = data[data['Number of Discharges']>=1000]
small = data[data['Number of Discharges']<1000]

# Getting the means

mean_small = small['Excess Readmission Ratio'].mean()
mean_large = large['Excess Readmission Ratio'].mean()

dif = mean_small - mean_large

# Getting the Variances

var_small = small['Excess Readmission Ratio'].var()
var_large = large['Excess Readmission Ratio'].var()

# Getting the Standard deviation for the difference sampling distribution

dif_std = np.sqrt((var_small/len(small)) + (var_large/len(large)))

# Getting the pvalue

z = dif / dif_std

# Assuming the CLT in the independence of the random variables..

pvalue = stats.norm.pdf(z)
pvalue
```

# 3

```python
print('Significance Level: 0.01 ')
print('P-value: ', pvalue)
```

Here we could see that the significance of this difference is statistic valid even with the threshold in 1%.

So there seems to be a difference between the small and large hospital in the excess readmission ratios.

# 4

To discuss this question it should be appropriate to set the confidence intervals for the difference so we can explore better the effect size.

```python
# Getting the thresholds

tr01 = stats.norm.ppf(.995) # 1% significance level
tr05 = stats.norm.ppf(.975) # 5% significance level
tr10 = stats.norm.ppf(.95) # 10% significance level


# Getting the margin of error

m01 = (dif_std*tr01) # 1% significance level
m05 = (dif_std*tr05) # 5% significance level
m10 = (dif_std*tr10) # 10% significance level

# Getting the intervals

int01 = (dif - (dif_std*tr01), dif + (dif_std*tr01)) # 1% significance level
int05 = (dif - (dif_std*tr05), dif + (dif_std*tr05)) # 5% significance level
int10 = (dif - (dif_std*tr10), dif + (dif_std*tr10)) # 10% significance level


print('Margin of Error for 1%: ' ,m01)
print('Margin of Error for 5%: ' ,m05)
print('Margin of Error for 10%: ' ,m10)

print('\nConfidence Interval for 1%: ' ,int01)
print('Confidence Interval for 5%: ' ,int05)
print('Confidence Interval for 10%: ' ,int10)
```

Here is clear for us that, statistically speaking the p-value indicates the significance for the difference with a very small probability. <br>

But is very importat to see that, even in the 1% margin of error, we have a very small pratical difference of 0.014. This indicates a very small proportion for a ratio that could go higher than 1. So in pratical terms we could say that this difference is irrelevant. We can confirm this by looking to the confidence interval for the 1% test, the maximum difference is 0.045, which is still ad very small number to consider relevant.
